Carilah gradient dari sebuah fungsi:
* $3x+4y=5$
* 5/4y - 3/4x
* y=1.25 - 0.75x

```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sy.Symbol('x')
y = sy.Symbol('y')
```

###**GRADIEN X**

```python
x = np.linspace(-50, 50)
```

```python
y = 1.25 - 0.75 * x
```

```python
Gradien_x = np.gradient(y)
Gradien_x
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(21,5))

ax[0].plot(x, y)
ax[0].set_title('x')

ax[1].plot(Gradien_x, y)
ax[1].set_title('Gradien_x')
plt.xlim(-2, 2)
```
