```python
import random
import matplotlib.pyplot as plt

trials = 50000
head_count = 0
tail_count = 0

for trial in range(0, trials):
  outcome = random.randint(0, 1)
  if outcome == 0:
    head_count += 1
  else:
    tail_count += 1

plt.figure(figsize=(7, 7))
plt.pie([head_count, tail_count], labels=['Head', 'Tail'])
plt.legend()
plt.show()
```
